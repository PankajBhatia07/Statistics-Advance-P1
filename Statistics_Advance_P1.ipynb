## **Q2.Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define the range of the discrete uniform distribution (e.g., a fair die: 1 to 6)
low, high = 1, 6

# Generate a discrete uniform distribution
x = np.arange(low, high + 1)
pmf = randint.pmf(x, low, high + 1)  # Probability mass function

# Plot the PMF
plt.bar(x, pmf, color='skyblue', edgecolor='black', alpha=0.7)
plt.xlabel("Random Variable (X)")
plt.ylabel("Probability P(X)")
plt.title("PMF of Discrete Uniform Distribution (1 to 6)")
plt.xticks(x)
plt.ylim(0, max(pmf) + 0.05)

# Show probability values on bars
for i, p in zip(x, pmf):
    plt.text(i, p + 0.01, f"{p:.2f}", ha='center', fontsize=12)

plt.show()
```

**Explanation:**
* randint.pmf(x, low, high + 1): Computes the PMF for a discrete uniform distribution.
* plt.bar(): Plots the probability values as a bar chart.
* Text Labels: Displays the probability on each bar for clarity.

## **Q4.Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters for the binomial distribution
n = 10   # Number of trials
p = 0.5  # Probability of success

# Simulate 10000 samples from the binomial distribution
samples = np.random.binomial(n, p, 10000)

# Plot histogram of the samples
plt.hist(samples, bins=np.arange(n+2)-0.5, density=True, alpha=0.7, color='skyblue', edgecolor='black', rwidth=0.8)

# Overlay the theoretical binomial probability mass function (PMF)
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)
plt.plot(x, pmf, 'ro-', markersize=8, label='Theoretical PMF')

# Labels and title
plt.xlabel("Number of Successes (X)")
plt.ylabel("Probability")
plt.title(f"Binomial Distribution Histogram (n={n}, p={p})")
plt.xticks(x)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show plot
plt.show()
```

**Explanation:**
* Simulates 10,000 trials of a binomial distribution using np.random.binomial(n, p, size).
* Plots a histogram to visualize the distribution.
* Overlays the theoretical PMF (red dots) using scipy.stats.binom.pmf().
* Uses density=True so the histogram matches the probability values.

## **Q6.Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define the range of the discrete uniform distribution
low, high = 1, 6  # Example: Rolling a fair 6-sided die

# Generate x values (possible outcomes)
x = np.arange(low, high + 1)

# Compute the CDF using scipy.stats
cdf = randint.cdf(x, low, high + 1)

# Plot the CDF
plt.step(x, cdf, where='mid', color='blue', label="CDF")
plt.scatter(x, cdf, color='red', zorder=3)  # Add markers for clarity

# Labels and title
plt.xlabel("Random Variable (X)")
plt.ylabel("Cumulative Probability P(X ‚â§ x)")
plt.title("CDF of Discrete Uniform Distribution (1 to 6)")
plt.xticks(x)
plt.yticks(np.round(cdf, 2))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.show()
```

## **Q7.Generate a continuous uniform distribution using NumPy and visualize it.**

**Explanation:**

* Plots a histogram of the generated samples.
* Overlays the theoretical PDF using scipy.stats.uniform.pdf().

## **Q8.Simulate data from a normal distribution and plot its histogram.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define the parameters of the normal distribution
mu, sigma = 0, 1  # Mean (Œº) and Standard Deviation (œÉ)

# Generate 10,000 random samples from the normal distribution
samples = np.random.normal(mu, sigma, 10000)

# Plot histogram of the samples
plt.hist(samples, bins=30, density=True, color='skyblue', edgecolor='black', alpha=0.7, label="Sampled Data")

# Theoretical Normal Distribution (PDF)
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)  # X values
pdf = norm.pdf(x, mu, sigma)  # Compute PDF
plt.plot(x, pdf, 'r-', lw=2, label="Theoretical PDF")

# Labels and title
plt.xlabel("Value (X)")
plt.ylabel("Density")
plt.title("Histogram of Normal Distribution (Œº=0, œÉ=1)")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def central_limit_theorem_demo(sample_size=30, num_samples=1000):
    """
    Demonstrates the Central Limit Theorem (CLT) using a non-normal distribution (Exponential).

    Parameters:
    sample_size (int): Size of each sample drawn from the non-normal distribution.
    num_samples (int): Number of samples to draw.
    """
    # Generate an exponentially distributed population (highly skewed, non-normal)
    population = np.random.exponential(scale=2, size=10000)

    # Collect sample means
    sample_means = [np.mean(np.random.choice(population, sample_size, replace=True)) for _ in range(num_samples)]

    # Plot original population distribution
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    sns.histplot(population, bins=50, kde=True, color='red', edgecolor='black', alpha=0.7)
    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.title("Original Exponential Distribution (Skewed)")

    # Plot distribution of sample means
    plt.subplot(1, 2, 2)
    sns.histplot(sample_means, bins=50, kde=True, color='blue', edgecolor='black', alpha=0.7)
    plt.xlabel("Sample Mean")
    plt.ylabel("Frequency")
    plt.title(f"Sampling Distribution (Sample Size = {sample_size})\n Approximates Normal")

    plt.tight_layout()
    plt.show()

# Run the CLT demonstration
central_limit_theorem_demo(sample_size=30, num_samples=1000)
```

## **Q15.Implement hypothesis testing using Z-statistics for a sample dataset.**


**Problem Statement:**
Suppose we want to test whether the mean weight of a population is 50 kg based on a sample dataset.
We assume a known population standard deviation (which allows us to use Z-statistics).

**Hypotheses:**

* Null Hypothesis (
ùêª
0
 ): The population mean is 50 kg.

* Alternative Hypothesis (
ùêª
ùê¥): The population mean is not 50 kg (two-tailed test)

```python
import numpy as np
import scipy.stats as stats

def z_test(sample, pop_mean, pop_std, alpha=0.05):
    """
    Performs a Z-test for a sample dataset.

    Parameters:
    sample (list or np.array): Sample dataset
    pop_mean (float): Hypothesized population mean (H0)
    pop_std (float): Known population standard deviation
    alpha (float): Significance level (default = 0.05)

    Returns:
    None
    """
    # Sample statistics
    sample_mean = np.mean(sample)
    sample_size = len(sample)

    # Calculate the Z-score
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))

    # Calculate p-value for two-tailed test
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    # Print results
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-score: {z_score:.2f}")
    print(f"P-value: {p_value:.5f}")

    # Conclusion
    if p_value < alpha:
        print("Reject the Null Hypothesis (H0): Significant difference detected.")
    else:
        print("Fail to Reject the Null Hypothesis (H0): No significant difference detected.")

# Example usage
np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(loc=51, scale=5, size=30)  # Sample data (mean=51, std=5, n=30)
population_mean = 50  # Hypothesized mean (H0)
population_std = 5    # Known population standard deviation

# Perform Z-test
z_test(sample_data, population_mean, population_std)
```

## **Q16.Create a confidence interval for a dataset using Python and interpret the result.**

```python
import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence=0.95):
    """
    Calculates the confidence interval for a given dataset.

    Parameters:
    data (list or np.array): Sample dataset
    confidence (float): Confidence level (default = 0.95 for 95% CI)

    Returns:
    tuple: Lower and upper bound of the confidence interval
    """
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Sample standard deviation (ddof=1 for unbiased estimate)
    n = len(data)

    # Calculate the margin of error using the t-distribution (for small samples)
    t_value = stats.t.ppf((1 + confidence) / 2, df=n-1)  # t-critical value
    margin_of_error = t_value * (sample_std / np.sqrt(n))

    # Compute confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound, sample_mean

# Example usage
np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(loc=70, scale=10, size=30)  # Sample of n=30, mean=70, std=10

# Calculate confidence interval
lower, upper, sample_mean = confidence_interval(sample_data)

# Print result
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
print("Interpretation: We are 95% confident that the true population mean lies within this range.")
```

## **Q22.Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.**

**Problem Statement:**

We generate a dataset from a normal distribution, then compute and interpret:

* **Range** = Max Value - Min Value

* **Interquartile Range (IQR)** = Q3 - Q1 (Difference between 75th and 25th percentiles)

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Generate a dataset from a normal distribution
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=50, scale=10, size=1000)  # Mean=50, Std=10

# Step 2: Calculate Range
data_range = np.max(data) - np.min(data)

# Step 3: Calculate Interquartile Range (IQR)
Q1 = np.percentile(data, 25)  # 25th percentile
Q3 = np.percentile(data, 75)  # 75th percentile
IQR = Q3 - Q1

# Step 4: Plot Box Plot for visualization
plt.figure(figsize=(8, 5))
sns.boxplot(data, color="skyblue")
plt.title("Box Plot of the Dataset")
plt.xlabel("Values")
plt.grid()

# Step 5: Print Results
print(f"Range of Dataset: {data_range:.2f}")
print(f"Interquartile Range (IQR): {IQR:.2f}")
print("Interpretation: IQR measures data spread within the middle 50%, while range shows full data spread.")

plt.show()
```

**Explanation:**

**Step 1:** Generate normally distributed data with np.random.normal().

**Step 2:** Compute Range using max - min.

**Step 3:** Compute Interquartile Range (IQR) using the 25th and 75th percentiles.

**Step 4:** Visualize the Box Plot to see data spread.

**Step 5:** Print and interpret the results.

## **Q24.Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.**

**What are Skewness & Kurtosis?**

**1.Skewness:** Measures asymmetry in the data distribution.

* Skewness = 0 ‚Üí Symmetric (Normal distribution).

* Skewness > 0 ‚Üí Right-skewed (Positive skew).

* Skewness < 0 ‚Üí Left-skewed (Negative skew).

**2.Kurtosis:** Measures tailedness (peak/sharpness) of the distribution.

* Kurtosis = 3 ‚Üí Normal distribution (Mesokurtic).

* Kurtosis > 3 ‚Üí Heavy tails (Leptokurtic).

* Kurtosis < 3 ‚Üí Light tails (Platykurtic).

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def compute_skewness_kurtosis(data):
    skewness = stats.skew(data)  # Measure of asymmetry
    kurtosis = stats.kurtosis(data, fisher=True)  # Excess kurtosis (subtracts 3)

    return skewness, kurtosis

# Step 1: Generate a dataset from a normal distribution
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=50, scale=10, size=1000)  # Mean=50, Std=10

# Step 2: Compute Skewness and Kurtosis
skewness, kurtosis = compute_skewness_kurtosis(data)

# Step 3: Plot the Histogram
plt.figure(figsize=(8, 5))
sns.histplot(data, bins=30, kde=True, color="skyblue", edgecolor="black")
plt.axvline(np.mean(data), color="red", linestyle="dashed", label=f"Mean={np.mean(data):.2f}")
plt.title("Histogram of Normally Distributed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.grid()

# Step 4: Print results
print(f"Skewness: {skewness:.4f}")
print(f"Kurtosis (Excess): {kurtosis:.4f}")
print("Interpretation:")
print("- Skewness ‚âà 0 ‚Üí Symmetric distribution.")
print("- Kurtosis ‚âà 0 ‚Üí Normal-like tail behavior.")

plt.show()
```

**Explanation:**

**Step 1:** Generate normally distributed data with mean = 50 and std dev = 10.

**Step 2:** Compute skewness (asymmetry) and kurtosis (tailedness).

**Step 3:** Plot the histogram to visualize the data shape.

**Step 4:** Print results and interpretation.
